```python
import sqlite3
import pandas as pd
import csv
import pymysql

import os
```

```python

con = pymysql.connect(
                        user='...', 
                        password= "SECRET", 
                        host='...', 
                        database='...')

```

```python
# creating cursor
cursor = con.cursor()

```

```python
'''Writes the contents of the cursor to a CSV'''
def writeCursor(cursor, fileName):
    rows = cursor.fetchall()
    column_names = [i[0] for i in cursor.description]
    fp = open(fileName, 'w')
    myFile = csv.writer(fp)
    myFile.writerow(column_names)
    myFile.writerows(rows)
    fp.close()
```

```python
cursor.execute("SELECT * FROM participants")


writeCursor(cursor, 'participants.csv')
```

```python
cursor.execute("SELECT * FROM responses")


writeCursor(cursor, 'responses.csv')
```
